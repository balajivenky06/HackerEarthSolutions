```python
t = int(input())
for _ in range(t):
    a = list(map(int, input().split()))
    i,j =0, len(a)-1
    while i != len(a) and j != -1:
        if a[j] > a[i]:
            print("j", j)
            print("i", i)
            print("a[j]", a[j])
            print("a[i]", a[i])
            print("2", end=" ")
            i += 1
        elif a[j] < a[i]:
            print("j", j)
            print("i", i)
            print("a[j]", a[j])
            print("a[i]", a[i])
            print("1", end=" ")
            j -= 1
        else:
            print("j", j)
            print("i", i)
            print("a[j]", a[j])
            print("a[i]", a[i])
            print("0", end=" ")
            i += 1
            j -= 1
```
