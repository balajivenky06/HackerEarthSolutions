```python
t = int(input())
for _ in range(t):
    n = int(input())
    arr = list(map(int, input().split()))
    print(arr[0], end=' ')
    for i in range(1, n):
        #print('arr[i-1]', arr[i-1])
        #print('arr[i]', arr[i])
        if arr[i-1] % arr[i] == 0:
            arr[i] =  arr[i-1]
        else:
            arr[i] = (arr[i-1]//arr[i] + 1) * arr[i]
        print(arr[i], end=' ')
    print()
```
