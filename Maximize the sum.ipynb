```python
from collections import Counter
t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    print("n:", n, "k:", k)
    a = list(map(int, input().split()))
    b = Counter(a)
    print(b)
    result = [val * key for key, val in b.items() if key > 0]
    result.sort(reverse=True)
    print(result)
    print(sum(result[0:k]))
```
