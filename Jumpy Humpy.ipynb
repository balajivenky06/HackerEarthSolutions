```python
n = int(input())
arr = [int(item) for item in input().split()]
stk = []
ans = arr.copy()
 
for i in range(n-1, -1, -1):
    while len(stk) > 0 and stk[-1][0] < arr[i]:
        stk.pop()
    if len(stk) > 0:
        ans[i] ^= ans[stk[-1][1]]
    stk.append([arr[i], i])
    print('stk[-1][0]', stk[-1][0])
    print('stk[-1][1]', stk[-1][1])
    print(ans)
    print(stk)
print(max(ans))
```
